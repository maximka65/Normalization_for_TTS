```python
import pandas as pd
from tqdm import tqdm
```

```python
data = pd.read_csv('ru_train.csv')
```

```python
data
```

```python
x_sentence_list = []
y_sentence_list = []
for i in tqdm(range(761436)):
    try:
        x_sentence = ' '.join(list(data.loc[data['sentence_id'] == i, 'before'])).encode('utf-8')
        y_sentence = ' '.join(list(data.loc[data['sentence_id'] == i, 'after'])).encode('utf-8')

        x_sentence_list.append(x_sentence)
        y_sentence_list.append(y_sentence)
    except:
        pass
```

```python
x_sentence_list
```

```python
final = [x_sentence_list, y_sentence_list]
```

```python
final
```

```python
import pickle

with open('train_list', 'wb') as f:
    pickle.dump(final, f)
```

```python
with open('train_list', 'rb') as f:
    la = pickle.load(f)
```
